IMPORT DEPENDICIES


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing

from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
# import california housing dataset

housing = fetch_california_housing()
print(housing)
```

```python
# loading the dataset to a pandas dataframe
house_price = pd.DataFrame(housing.data,columns = housing.feature_names)
# print first 5 rows of our dataset
house_price.head()
```

```python
# add the target (house price) column  to the dataframe
house_price['price'] = housing.target

house_price.head()
```

```python
# checking the number of rows and columns in data frame
house_price.shape
```

```python
# check for missing values
house_price.isnull().sum()
```

```python
# getting statistical measure
house_price.describe()
```

understanding the correlation between various features in datasets
1. postive correlation +1
2. negative correlation -1


```python
correlation = house_price.corr()
```

```python
# contructing a heatmap to understand the correlation
plt.figure(figsize=(10,10 ))
sns.heatmap(correlation, cbar=True, square=True, fmt='.2f', annot=True, annot_kws={'size':8}, cmap='Greens' )
```

splitting the data into target and features

```python
X = house_price.drop(['price'],axis=1)
Y = house_price['price']
print(X)
print(Y)
```

```python
#splitting the data into training data and test data
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state = 3)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
# training the model,using XGBoost Regression
# loading the model
model = XGBRegressor()
# Training the model with X _train
model.fit(X_train,Y_train)

```

```python
#Evaluation
# prediction on training data
tr_prediction = model.predict(X_train)
# R square error
score_1 = metrics.r2_score(Y_train,tr_prediction)
# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_train,tr_prediction)
print("R square error : ",score_1)
print("Mean Absolute Error : ",score_2)


```

```python
# Visualing the actual prices and predicted prices
plt.scatter(Y_train,tr_prediction)
plt.xlabel("Actual prices")
plt.ylabel("Predicted prices")
plt.title("Actual prices vs Predicted prices")
plt.show()
```

```python
# prediction on testing data
te_prediction = model.predict(X_test)
# R square error
score_1 = metrics.r2_score(Y_test,te_prediction)
# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_test,te_prediction)
print("R square error : ",score_1)
print("Mean Absolute Error : ",score_2)
```

```python
# making a Prediction system
input_data = ( 8.3252,45,6.97,1.810,322.0,2.555556,37 ,-122)
# converting the input to numpy array
np_array = np.asarray(input_data)
# reshape it for one instance
reshaped = np_array.reshape(1,-1)
prediction = model.predict(reshaped)
print(f'The house price is :  { int(prediction[0])} thousand dollars')
```
